```python
def wild_augmentation(img):

    # green gaussian noise base            
    base = np.random.normal(0,1,img.size)
    base = base.reshape(256,256,3).astype('uint8')

    # blur base
    base = ndimage.gaussian_filter(base, sigma=3)

    for i in range(256):
        for j in range(256):
            if np.all((img[i][j] <= 50)):
                np.clip(base[i][j] + [0,50,0], 0, 255, img[i][j]) 

    return img
```
